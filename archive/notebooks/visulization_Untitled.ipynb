```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 读取数据并清理 '±' 部分
file_path = 'letter'
data = pd.read_csv(file_path+'_res.csv')

# 清理数据，去除每个 metrics 列中的 '±' 部分
metrics = ['Accuracy', 'Precision', 'Recall', 'F1']
#for metric in metrics:
#    data[metric] = data[metric].str.split('±').str[0].astype(float)

# 将数据转换为长格式以便使用 seaborn 作图
data_long = data.melt(id_vars='Models', value_vars=metrics, var_name='Metrics', value_name='Value(%)')

# 设置 seaborn 样式和配色方案
sns.set(style="whitegrid")
palette = sns.color_palette("pastel")
plt.rcParams['figure.dpi']=1200
plt.rcParams['font.family']='Times New Roman'

# 绘图
plt.figure(figsize=(12, 5))
sns.barplot(x='Metrics', y='Value(%)', hue='Models', data=data_long, palette=palette)

# 调整图例位置到右侧
plt.legend(title="Models", loc='center left', bbox_to_anchor=(1, 0.5), fontsize=14, title_fontsize=18)

plt.xlabel('')
plt.ylabel('Values', fontsize=28)
plt.xticks(fontsize=26)
plt.yticks(fontsize=26)

# 调整 y 轴范围以突出差异
plt.ylim(0, 1.02)

# 添加标题和格式
# plt.title('Comparison of Models across Metrics')
plt.tight_layout()

plt.savefig(file_path+".eps", format="eps", dpi=1200)
plt.show()
```
